```python
import datetime

# Memory / database for tickets
ticket_db = []

# Predefined support actions
solutions = {
    "login": "Please reset your password using the 'Forgot Password' option.",
    "payment": "Check if your card is active or try another payment method.",
    "slow": "Try clearing browser cache or check your internet connection.",
    "error": "Please reinstall or update the app to the latest version."
}

def support_agent():
    print("ü§ñ Welcome to Customer Support Agent!")

    user_id = input("Enter your user ID: ").strip()
    issue = input("Describe your issue briefly: ").lower().strip()

    timestamp = datetime.datetime.now().isoformat()

    # Save issue in DB
    ticket_db.append({"user_id": user_id, "issue": issue, "time": timestamp})

    # Identify solution
    found_solution = None
    for keyword, solution in solutions.items():
        if keyword in issue:
            found_solution = solution
            break

    print("\nüìå Support Response:")
    if found_solution:
        print(found_solution)
    else:
        print("We have forwarded your issue to a human expert. Please wait 24 hours.")

    print("\nüìù Ticket Recorded!")
    print(f"User: {user_id}")
    print(f"Issue: {issue}")
    print(f"Time: {timestamp}")

def view_all_tickets():
    print("\nüìö ALL TICKETS LOGGED:")
    for idx, t in enumerate(ticket_db, start=1):
        print(f"{idx}. {t['user_id']} | {t['issue']} | {t['time']}")

# --------- MAIN LOOP ----------
while True:
    print("\n1Ô∏è‚É£ Raise New Issue")
    print("2Ô∏è‚É£ View Issue Records")
    print("3Ô∏è‚É£ Exit")

    choice = input("Select an option: ").strip()

    if choice == "1":
        support_agent()
    elif choice == "2":
        view_all_tickets()
    elif choice == "3":
        print("Goodbye! üëã")
        break
    else:
        print("Invalid choice. Try again!")
```
